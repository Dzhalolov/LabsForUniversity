```python
from math import factorial as fact
def coef(h, y, k):
    b = 0
    for i in range(k+1):
        b += (-1) ** (k-i) * y[i] / (fact(i) * fact(k-i) * h ** k)
    return b


def multiply(i, x, x_new):
    result = 1
    for index in range(i):
        result *= x_new - x[index]
    return result


def newton(x, x_new, n, y):
    P = 0
    h = x[1] - x[0]
    for i in range(n):
        P += coef(h, y, i) * multiply(i, x, x_new)
    return P
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = -10
b = 10
n = 8
x = np.linspace(a, b, n)
y = x ** 2
plt.plot(x, y)
print("x: {} \ny: {}".format(x, y))
x_new = 1.5
y_new = newton(x, x_new, n, y)
plt.plot(x_new, y_new, color='green', marker='o', linestyle='dashed')
```

```python
def proccesing(x, n, y):
    result = []
    for i in range(len(x)-1): 
        result.append([(x[i]+x[i+1])/2, newton(x, (x[i]+x[i+1])/2, n, y)])
    return result
```

```python
x_new = proccesing(x, n, y)
```

```python
x_new = np.array(x_new)
```

```python
plt.plot(x_new[:, 0], x_new[:, 1])
plt.plot(x, y)
```

```python
x_all = list(np.c_[x[:-1], x_new[:, 0]].reshape(-1, ))
x_all.append(x[-1])
y_all = list(np.c_[y[:-1], x_new[:, 1]].reshape(-1, ))
y_all.append(y[-1])
```
