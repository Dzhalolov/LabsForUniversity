```python
from math import factorial as fact
def coef(h, y, k):
    b = 0
    print("y is {}, k is {}, h is {}".format(y, k, h))
    for i in range(k):
        b += (-1) ** (k-i) * y[i] / (fact(i) * fact(k-i) * h ** k)
    print("b is [{}]".format(b))
    return b


def multiply(i, x, x_new):
    result = 1
    for index in range(i):
        result *= (x_new - x[index])
    return result


def newton(x, x_new, n, y):
    P = 0
    h = x[1] - x[0]
    for i in range(n):
        P += coef(h, y, i) * multiply(i, x, x_new)
    return P
```
