```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 200

from matplotlib import rcParams
rcParams['figure.figsize'] = 14, 12
rcParams['font.size'] = 16
rcParams['axes.labelsize'] = 14
rcParams['xtick.labelsize'] = 13
rcParams['ytick.labelsize'] = 13
rcParams['legend.fontsize'] = 15

import seaborn as sns
sns.set_style("whitegrid")
source_data = pd.read_csv('2.csv', sep=';') #, index_col=[0], parse_dates=[0]
source_data.head()
```

```python
source_data.dow.unique()
```

```python
clean_data = source_data.copy()
```

```python
clean_data['dow'] = source_data['dow']
clean_data['season'] = source_data['season'].replace({'Зима': 0, 'Весна': 1, 'Лето': 2, 'Осень': 3})
clean_data['working_day'] = source_data['working_day'].replace({True: 1, False: 0})
clean_data['to_over_margin'] = source_data['to_over_margin'].replace({True: 1, False: 0})
clean_data['action'] = source_data['action'].replace({True: 1, False: 0})
clean_data['drug_avail'] = source_data['drug_avail'].replace({True: 1, False: 0})

clean_data.head()
```

```python
known_client_ids = [i for i in range(1,18)]
known_client_ids.remove(12)
known_client_ids.remove(13)
known_client_ids.remove(15)
known_client_ids.remove(16)

distr_bought_columns = ['client_distr{}_bought'.format(i) for i in known_client_ids]
distr_bought_columns.append('client_own_bought')

distr_price_columns = ['client_distr{}_price'.format(i) for i in known_client_ids]
distr_price_columns.append('client_own_price')

clean_data = clean_data.drop(columns=['upper_limit', 'to_over_margin', 'first_price', 'lower_limit'])
```

```python
# заполнять nan через -1.0 не пойдёт...

clean_data[distr_bought_columns] = clean_data[distr_bought_columns].fillna(0.0)

# будем заполнять цену предыдущими известными значениями

# 
clean_data[distr_price_columns] = clean_data[distr_price_columns].fillna(method='bfill').fillna(method='ffill')

#clean_data[distr_price_columns] = clean_data[distr_price_columns].fillna(-1.0)
#clean_data['lower_limit'] = clean_data['lower_limit'].fillna(-1.0)
#clean_data['first_price'] = clean_data['first_price'].fillna(-1.0)
```

```python
normalize_data = clean_data.copy()
normalize_data["mnth_sin"] = np.sin((normalize_data.mnth-1)*(2.*np.pi/12))
normalize_data["mnth_cos"] = np.cos((normalize_data.mnth-1)*(2.*np.pi/12))
normalize_data["dow_sin"] = np.sin((normalize_data.dow-1)*(2.*np.pi/7))
normalize_data["dow_cos"] = np.cos((normalize_data.dow-1)*(2.*np.pi/7))
normalize_data["m_in_q_sin"] = np.sin((normalize_data.m_in_q-1)*(2.*np.pi/3))
normalize_data["m_in_q_cos"] = np.cos((normalize_data.m_in_q-1)*(2.*np.pi/3))
# лучше ohe для season
normalize_data["season_sin"] = np.sin((normalize_data.season-1)*(2.*np.pi/4))
normalize_data["season_cos"] = np.cos((normalize_data.season-1)*(2.*np.pi/4))

normalize_data.head()
```

```python
normalize_data.describe()
```

```python
# остальное масштабируем standard scaler, предполагается что колонки имеют нормальное распределение, хотя это, скорее всего, не так

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
# для каких столбцов? Всех, содержащих own или distr

standardize_list_col = [col for col in normalize_data.columns if ('distr' in col) | ('own' in col)]

col_to_scaler = dict()

#print(standardize_list_col)

for col in standardize_list_col:
    scaler = StandardScaler()
    values=normalize_data[col].values.reshape(-1,1)
    scaler.fit(values)
    #print(col)
    #print(values)
    transformed=scaler.transform(values).reshape(1,-1)[0]
    #print(transformed)
    #print(scaler.inverse_transform(transformed.reshape(-1,1)))
    normalize_data[col] = transformed
    col_to_scaler[col] = scaler
  
normalize_data.head()
print(col_to_scaler)
```

```python
normalize_data_idx = ((normalize_data['client_id'] == '90060000000002FE') & (normalize_data['drug_ag_id'] == '880000000000EEB7'))
predicted_data = normalize_data[normalize_data_idx]
predicted_data = predicted_data.drop(columns=['drug_name', 'drug_ag_id', 'client_id', 'mnth', 'dow', 'm_in_q', 'season'])
predicted_data.describe()
```

```python
split_test_date = '2019-06-01'
split_val_date = '2019-07-01'
train_data = predicted_data.loc[predicted_data.dt < split_test_date].copy().reset_index(drop=True)
test_data = predicted_data.loc[(predicted_data.dt >= split_test_date) & (predicted_data.dt < split_val_date)].copy().reset_index(drop=True)
val_data = predicted_data.loc[predicted_data.dt >= split_val_date].copy().reset_index(drop=True)
```

```python
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, LSTM, TimeDistributed
from keras.optimizers import SGD
from keras import backend
from livelossplot import PlotLossesKeras
import tensorflow as tf
opt = SGD(lr=0.0001)
from keras_radam import RAdam
# Сколько временных шагов используется на входе (дней)
n_timesteps_in = 30

def rmse(y_true, y_pred):
	return backend.sqrt(backend.mean(backend.square(y_pred - y_true), axis=-1))
# Построим LSTM модель
def build_model():
    # Предсказывать будем цены и объёмы
    # в качестве признаков оставляем все атрибуты 
    n_features = len(predicted_data.columns) - 1
    print('Количество признаков: {}'.format(n_features))       
    
    # Предсказываем 1 день и всего n_features_out признаков на этот день
    n_features_out = len(standardize_list_col)
    
    print('Количество предсказываемых признаков: {}'.format(n_features_out))
    print(standardize_list_col)
    model = Sequential()

    model.add(LSTM(250, input_shape=(n_timesteps_in,n_features), return_sequences=False))
#     model.add(Dropout(0.4)) #регуляризация чтобы сеть не переобучилась, 0.2 большая...
    #model.add(LSTM(250, return_sequences=False, activation='relu'))
    #model.add(Dropout(0.4))
    # добавляем второй слой LSTM
    # абсолютная эвристика с количеством скрытых юнитов
    #model.add(LSTM(1000, return_sequences=False))
    #model.add(Dropout(0.2))    
    
    model.add(Dense(1))

    model.compile(loss="mse", optimizer=RAdam(total_steps=10000, warmup_proportion=0.1, min_lr=1e-5))
    return model
  
 
model = build_model()

print(model.summary())
RAdam()
```

```python
def training_data_generator(df, i, batch_size = 2):
  in_features = df.drop(columns='dt')
  out_features = df[standardize_list_col[i]]
  while True:
    current_batch = 0
    batch_input = []
    batch_output = []
    for i in range(n_timesteps_in + 1, len(df)):
      if current_batch == 0:      
          if len(batch_input) > 0:
            yield(np.array(batch_input), np.array(batch_output))

          batch_input = []
          batch_output = []

      batch_input.append(in_features.iloc[i-n_timesteps_in-1:i-1].values)
      batch_output.append(out_features.iloc[i-1:i].values[0])    

      current_batch = 0 if current_batch == batch_size - 1 else (current_batch + 1)

#     if len(batch_input) > 0:
#           yield(np.array(batch_input), np.array(batch_output))
      
# for x,y in training_data_generator(test_data.reset_index(drop=True)):  
#   print(x)
#   print(y)
#   break
# test_data.reset_index(drop=True).head(n_timesteps_in+5)   
```
